```python
from data_utils import load_images

# a123f007-d4ff-428b-9953-9b079b5ab2b1
validation_image_dataset = load_images("validation")
print(f"Loaded {len(validation_image_dataset)} validation images.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Select a sample image by id
img, label = validation_image_dataset.get_image_by_id(
    "1baab3be-fb1b-4ba1-9a5d-1e50217749f9"
)

r = img[:256]
g = img[256:512]
b = img[512:]

# Plot in a 2x2 grid: 2 on the first row, 1 on the second
fig, axes = plt.subplots(2, 2, figsize=(15, 7), sharey=True)
channels = [r, g, b]
colors = ["red", "green", "blue"]
titles = ["Red Channel Histogram", "Green Channel Histogram", "Blue Channel Histogram"]

# Flatten axes for easy indexing
axes_flat = axes.flatten()

for i, (channel, color, title) in enumerate(zip(channels, colors, titles)):
    ax = axes_flat[i]
    ax.bar(range(256), channel, color=color, alpha=0.7)
    ax.set_title(title, fontsize=18)
    ax.set_xlabel("Pixel Value", fontsize=16)
    ax.set_xlim(0, 255)
    ax.grid(axis="y", linestyle="--", alpha=0.5)
axes_flat[0].set_ylabel("Frequency", fontsize=16)

# Hide the unused subplot (bottom right)
axes_flat[3].axis("off")

plt.tight_layout()
plt.show()
```
